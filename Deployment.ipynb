####Push Image to Docker Hub :

```python
## Tag local image :
docker tag <image-name> <username>/<repo-name>:<tag-name>

<image-name>   --> name of the image on local
<username>     --> DockerHub username
<repo-name>    --> DockerHub repo name
<tag-name>     --> DockerHub tag name

docker tag demo-img dattaugale/31jan:hr-project
```

```python
## Push image to dockerhub :
docker push <username>/<repo-name>:<tag-name>
```
